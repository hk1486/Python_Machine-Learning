### 기상 데이터 분석하기

```python
input_file = 'data.csv'
output_file = 'tem10y.csv'
```

```python
# CSV 파일을 한 줄씩 읽어오기
with open(input_file, "rt", encoding='EUC-KR') as fr:
    lines = fr.readlines()
```

```python
# 기존의 데이터를 분리해서 가공하기
lines = ['연,월,일,기온,품질,균질\n']+lines[5:] # 기존헤더 줄을 제거하고 새로운 헤더를 추가
lines = map(lambda v: v.replace('/',','), lines) # / -> , 로 바꿔줌
result = "".join(lines).strip()
print(result)
```

```python
with open(output_file, 'wt', encoding='utf-8') as fw:
    fw.write(result)
    print('저장되었습니다.')
```

### 평균 기온 구하기
- 10년 동안의 데이터를 집계해서 각 연월의 평균 기온을 구합니다.

```python
# array는 []를 쓰는 대표적인 타입입니다. []는 배열을 선언 및 초기화할 때 배열의 원소에 접근할때 사용합니다.

# tuple은 ()를 쓰는 대표적인 타입입니다. ()는 튜플을 선언 및 초기화 할 때 사용합니다.
# []는 튜플의 원소에 접근할 때 사용합니다.

# dictionary는 {}를 쓰는 대표적인 타입입니다. {}는 딕셔너리를 선언 및 초기화할 때 사용합니다.
# []는 key에 대응하는 value를 활당하거나, value에 접근할 때 쓰입니다.

md = {}
# 데이터 프레임 자료형의 데이터를 한줄씩 처리하고 싶을 때 for ~ df.iterrows() 매소드 사용
for i, row in df.iterrows():
    m, d, v = (int(row['월']), int(row['일']), float(row['기온']))
    key = str(m) + "/" + str(d)
    if not(key in md): md[key] = []
    md[key] += [v]

avs = {}
for key in md:
    v = avs[key] = sum(md[key]) / len(md[key])
    print("{0} : {1}".format(key, v))
```

```python
# 월별 평균 기온 구하기
# 판다스 모듈의 groupby() 매소드를 사용해 간단하게 집계가능

g = df.groupby(['월'])['기온']
gg = g.sum() / g.count()
print(gg)

import matplotlib.pyplot as plt
gg.plot()
plt.savefig('tmp-month-avg.png') # 시각화 그래프를 이미지로 저장가능
plt.show()

"""
md = {}
# 데이터 프레임 자료형의 데이터를 한줄씩 처리하고 싶을 때 for ~ df.iterrows() 매소드 사용
for i, row in df.iterrows():
    m, d, v = (int(row['월']), int(row['일']), float(row['기온']))
    key = str(m)
    if not(key in md): md[key] = []
    md[key] += [v]

avs = {}
for key in md:
    v = avs[key] = sum(md[key]) / len(md[key])
    print("{0} : {1}".format(key, v))
"""

# 출력
"""
월
1     평균기온
2     평균기온
3     평균기온
..    평균기온
"""
```

```python
# 기온이 30도 넘는 날 구하기
# 평균 기온이 30도를 넘는날이 한해에 몇번 있었는지 확인

# 기온이 30도를 넘는 데이터 확인
hot_bool = (df['기온'] > 30)
# print(hot_bool)
hot = df[hot_bool]
cnt = hot.groupby(['연'])['연'].count()
print(cnt)
cnt.plot()
plt.savefig('tem-over30.png')
plt.show()

# 출력
"""
연
2006  몇번
2007  몇번
...   몇번
"""
```
